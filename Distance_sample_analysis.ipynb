```python
import numpy as np
import matplotlib.pyplot as plt
from likelihood import likelihood
```

```python
distance_samples = np.loadtxt('/Users/zero4zax/Downloads/distance_samples.dat')
distance_samples = distance_samples[ : , 1:]
# print(distance_samples)
```

```python
mean_trials = np.linspace(90,110,200)
std_trials = np.linspace(10,30,200)

mean_array = []
std_array = []
likelihood_array = []
for mean in mean_trials:
    for std in std_trials:
        mean_array.append(mean)
        std_array.append(std)
        likelihood_array.append(likelihood(distance_samples, mean, std))
        
likelihood_array = np.array(likelihood_array)
```

```python
mean_values, std_values = np.meshgrid(mean_trials, std_trials)


likelihood_array = likelihood_array.reshape((int(np.sqrt(likelihood_array.size)), int(np.sqrt(likelihood_array.size))))
heatmap = plt.pcolormesh(mean_values, std_values, likelihood_array, cmap = 'hot')
plt.colorbar(heatmap)
plt.title('Heatmap for mean and std. dev')
plt.xlabel("Mean values")
plt.ylabel("Std dev values")
plt.show()
```
